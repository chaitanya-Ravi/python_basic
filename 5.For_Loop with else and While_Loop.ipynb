#  While Loop

```python
a = 1
while a < 6:
    print(a)
    a = a + 1
```

```python
a = 1
while a <= 6:
    print(a)
    a = a + 1
```

```python
a = 1
while a < 5:
    print(a)
    if a == 4:
        break
    a = a + 1
```

```python
a = 1
while a<5:    ##  T T T
    print(a)  ##  1 2 3
    if a == 3:##  F F T (if we are using continue keyword it will again go to while loop condition infinitely )
        continue
    a = a+1   ##  2 3
```

```python
a = 1
while a<5:    ##  T T T
    print(a)  ##  1 2 3
    a = a+1   ##  2 3 4
    if a == 3:##  F T F
        continue
    
```

```python
n = 5
for i in range(0,n): ### this loop will work as outer loop and create row values
    for j in range(0,i+1): ### this loop will work as Inner loop and create column values
        print("*",end = "")
#         print("sudh ", end= "")
    print("\r")
```

```python
n = 5
for i in range(0,n): ### this loop will work as outer loop and create row values
    for j in range(0,i+1): ### this loop will work as Inner loop and create column values
        print(j+1, end= "")
    print("\r")
```

```python
n = 5
for i in range(n,0,-1):
    for j in range(0,i):
        print("*",end = "")
    print("\r")
    
```

```python
n = 5
for i in range(n,0,-1):
    for j in range(1,i+1):
        print(j,end = "")
    print("\r")
    
```

```python
rows = 6
number = 1

for i in range(1, rows+1):
    for j in range(1, i+1):
        print(number, end=" ")
        number += 1
    print()
```
